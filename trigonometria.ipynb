# Trigonometria

  Para trabalhar com operações e funções trigonométricas, foram desenvolvidos algoritmos para os mesmos. Em algumas funções há a presença dos casos notáveis já definidos, uma vez que a precisão dos algoritmos é limitada devido ao número máximo de interações. Também foram importadas constantes e funções dos módulos "constantes" e "aritmetica" já desenvolvidos.
  
  Primeiramente, foram definidas as operações de conversão entre graus e radianos, e em seguida a de normalização de ângulos, que normaliza ângulos fora de [-pi; pi) para esse intervalo.
  Foram desenvolvidas as funções trigonométricas seno, cosseno, tangente, cotangente, secante e cossecante; isso por meio da série de Taylor e das relações trigonométricas. 
  Também foram desenvolvidas as funções inversas, sendo arco seno, arco cosseno, arco tangente, arco cotangente, arco secante e arco cossecante, isso por meios série de Gregory-Leibniz com redução de argumento e pelas relações entre essas funções trigonométricas. 
  Por fim, foram desenvolvidas as funções trigonométricas hiperbólicas seno hiperbólico, cosseno hiperbólico, tangente hiperbólica, cotangente hiperbólica, secante hiperbólica e cossecante hiperbólica, todas por meio da definição exponencial e das relações trigonométricas entre elas.

  Para finalizar, adicionou-se a linha de código "%%writefile trigonometria.py" para converter o código em um módulo exportável para ser usado em outros módulos e no uso final. Pelo mesmo motivo, o código teve que ser colocado inteiramente em uma única célula de código, pois o comando usado para isso exporta códigos de apenas uma célula para isso.

```python
%%writefile trigonometria.py

from constantes import pi, tolerancia_de_erro, maximo_de_interacoes
from aritmetica import piso, n_raiz

#Conversão de graus para radianos
def radianos(gr: float) -> float:

    return gr * pi / 180.0


#Conversão de radianos para graus
def graus(rad: float) -> float:

    return rad * 180.0 / pi


#Normalização de ângulos para o intervalo [-pi; pi]
def normalizacao_angular(x: float) -> float:

    dois_pi = 2.0 * pi

    k = piso((x + pi) / dois_pi)
    return x - k * dois_pi


#Funções trigonométricas
#Seno
def sin(x: float) -> float:

    x = normalizacao_angular(x)

    #Pontos notáveis
    if abs(x) < tolerancia_de_erro:           return 0.0
    if abs(x - pi/6) < tolerancia_de_erro:    return 0.5
    if abs(x - pi/4) < tolerancia_de_erro:    return 2**0.5 / 2
    if abs(x - pi/3) < tolerancia_de_erro:    return 3**0.5 / 2
    if abs(x - pi/2) < tolerancia_de_erro:    return 1.0
    if abs(x + pi/6) < tolerancia_de_erro:    return -0.5
    if abs(x + pi/4) < tolerancia_de_erro:    return -(2**0.5 / 2)
    if abs(x + pi/3) < tolerancia_de_erro:    return -(3**0.5 / 2)
    if abs(x + pi/2) < tolerancia_de_erro:    return -1.0
    if abs(abs(x) - pi) < tolerancia_de_erro: return 0.0

    resultado = 0.0
    termo = x
    raiz_x = x * x

    for k in range(1, maximo_de_interacoes):
        resultado += termo

        termo *= -raiz_x / ((2 * k) * (2 * k + 1))
        if abs(termo) < tolerancia_de_erro * 1e-8:
            break

    return resultado


#Cosseno
def cos(x: float) -> float:

    x = normalizacao_angular(x)

    #Pontos notáveis
    if abs(x) < tolerancia_de_erro:           return 1.0
    if abs(x - pi/6) < tolerancia_de_erro:    return 3**0.5 / 2
    if abs(x - pi/4) < tolerancia_de_erro:    return 2**0.5 / 2
    if abs(x - pi/3) < tolerancia_de_erro:    return 0.5
    if abs(x - pi/2) < tolerancia_de_erro:    return 0.0
    if abs(x + pi/6) < tolerancia_de_erro:    return 3**0.5 / 2
    if abs(x + pi/4) < tolerancia_de_erro:    return 2**0.5 / 2
    if abs(x + pi/3) < tolerancia_de_erro:    return 0.5
    if abs(x + pi/2) < tolerancia_de_erro:    return 0.0
    if abs(abs(x) - pi) < tolerancia_de_erro: return -1.0

    resultado = 0.0
    termo = 1.0
    raiz_x = x * x

    for k in range(1, maximo_de_interacoes):
        resultado += termo
        termo *= -raiz_x / ((2 * k - 1) * (2 * k))
        if abs(termo) < tolerancia_de_erro * 1e-8:
            break

    return resultado


#Tangente
def tan(x: float) -> float:

    if abs(cos(x)) < tolerancia_de_erro:
        raise ZeroDivisionError("Tangente é indefinida quando cosseno ≈ 0.")
    return sin(x) / cos(x)


#Cotangente
def cot(x: float) -> float:

    if abs(sin(x)) < tolerancia_de_erro:
        raise ZeroDivisionError("Cotangente é indefinida quando seno ≈ 0.")
    return cos(x) / sin(x)


#Secante
def sec(x: float) -> float:

    if abs(cos(x)) < tolerancia_de_erro:
        raise ZeroDivisionError("Secante é indefinida quando cosseno ≈ 0.")
    return 1.0 / cos(x)


#Cossecante
def csc(x: float) -> float:

    if abs(sin(x)) < tolerancia_de_erro:
        raise ZeroDivisionError("Cossecante é indefinida quando seno ≈ 0.")
    return 1.0 / sin(x)


#Funções trigonométricas inversas
#Arco tangente
def arctan(x: float) -> float:
    if abs(x)        < tolerancia_de_erro: return 0.0
    if abs(x - 1.0)  < tolerancia_de_erro: return pi / 4.0
    if abs(x + 1.0)  < tolerancia_de_erro: return -pi / 4.0
    if abs(x - 3**0.5)   < tolerancia_de_erro: return pi / 3.0
    if abs(x + 3**0.5)   < tolerancia_de_erro: return -pi / 3.0
    if abs(x - 3**0.5/3) < tolerancia_de_erro: return pi / 6.0
    if abs(x + 3**0.5/3) < tolerancia_de_erro: return -pi / 6.0
    if x > 1.0:
        return pi / 2.0 - arctan(1.0 / x)
    if x < -1.0:
        return -pi / 2.0 - arctan(1.0 / x)
    if abs(x) > 0.5:
        u = (x - 1.0) / (x + 1.0)
        return pi / 4.0 + arctan(u)
    x2 = x * x
    termo = x
    resultado = 0.0
    sinal = 1
    for k in range(maximo_de_interacoes):
        resultado += sinal * termo / (2 * k + 1)
        termo *= x2
        sinal = -sinal
        if abs(termo) / (2 * k + 3) < tolerancia_de_erro * 1e-8:
            break
    return resultado

#Arco seno
def arcsin(x: float) -> float:

    #Pontos notáveis
    if abs(x - 1.0)  < tolerancia_de_erro: return pi / 2.0
    if abs(x + 1.0)  < tolerancia_de_erro: return -pi / 2.0
    if abs(x)        < tolerancia_de_erro: return 0.0
    if abs(x - 0.5)  < tolerancia_de_erro: return pi / 6.0
    if abs(x + 0.5)  < tolerancia_de_erro: return -pi / 6.0
    if abs(x - 2**0.5/2) < tolerancia_de_erro: return pi / 4.0
    if abs(x + 2**0.5/2) < tolerancia_de_erro: return -pi / 4.0
    if abs(x - 3**0.5/2) < tolerancia_de_erro: return pi / 3.0
    if abs(x + 3**0.5/2) < tolerancia_de_erro: return -pi / 3.0

    if abs(x) > 1.0 + tolerancia_de_erro:
        raise ValueError("Arco seno indefinido, o domínio dessa função é [−1, 1].")
    if x >= 1.0:
        return pi / 2.0
    if x <= -1.0:
        return -pi / 2.0

    return arctan(x / n_raiz(1 - x * x, 2))


#Arco cosseno
def arccos(x: float) -> float:

    #Pontos notáveis
    if abs(x - 1.0)  < tolerancia_de_erro: return 0.0
    if abs(x + 1.0)  < tolerancia_de_erro: return pi
    if abs(x)        < tolerancia_de_erro: return pi / 2.0
    if abs(x - 0.5)  < tolerancia_de_erro: return pi / 3.0
    if abs(x + 0.5)  < tolerancia_de_erro: return 2.0 * pi / 3.0
    if abs(x - 2**0.5/2) < tolerancia_de_erro: return pi / 4.0
    if abs(x + 2**0.5/2) < tolerancia_de_erro: return 3.0 * pi / 4.0
    if abs(x - 3**0.5/2) < tolerancia_de_erro: return pi / 6.0
    if abs(x + 3**0.5/2) < tolerancia_de_erro: return 5.0 * pi / 6.0

    if abs(x) > 1.0 + tolerancia_de_erro:
        raise ValueError("Arco cosseno indefinido, o domínio dessa função é [−1, 1].")
    return pi / 2.0 - arcsin(x)


#Arco cotangente
def arccot(x: float) -> float:

    #Pontos notáveis
    if abs(x)        < tolerancia_de_erro: return pi / 2.0
    if abs(x - 1.0)  < tolerancia_de_erro: return pi / 4.0
    if abs(x + 1.0)  < tolerancia_de_erro: return 3.0 * pi / 4.0
    if abs(x - 3**0.5)   < tolerancia_de_erro: return pi / 6.0
    if abs(x + 3**0.5)   < tolerancia_de_erro: return 5.0 * pi / 6.0
    if abs(x - 3**0.5/3) < tolerancia_de_erro: return pi / 3.0
    if abs(x + 3**0.5/3) < tolerancia_de_erro: return 2.0 * pi / 3.0

    return pi / 2.0 - arctan(x)


#Arco secante
def arcsec(x: float) -> float:

    #Pontos notáveis
    if abs(x - 1.0)  < tolerancia_de_erro: return 0.0
    if abs(x + 1.0)  < tolerancia_de_erro: return pi
    if abs(x - 2.0)  < tolerancia_de_erro: return pi / 3.0
    if abs(x + 2.0)  < tolerancia_de_erro: return 2.0 * pi / 3.0
    if abs(x - 2**0.5) < tolerancia_de_erro: return pi / 4.0
    if abs(x + 2**0.5) < tolerancia_de_erro: return 3.0 * pi / 4.0
    if abs(x - 2*3**0.5/3) < tolerancia_de_erro: return pi / 6.0
    if abs(x + 2*3**0.5/3) < tolerancia_de_erro: return 5.0 * pi / 6.0
        
    if abs(x) < 1.0 - tolerancia_de_erro:
        raise ValueError("Arco secante indefinido, o domínio dessa função é |x| ≥ 1.")
    return arccos(1.0 / x)


#Arco cossecante
def arccsc(x: float) -> float:

    #Pontos notáveis
    if abs(x - 1.0)  < tolerancia_de_erro: return pi / 2.0
    if abs(x + 1.0)  < tolerancia_de_erro: return -pi / 2.0
    if abs(x - 2.0)  < tolerancia_de_erro: return pi / 6.0
    if abs(x + 2.0)  < tolerancia_de_erro: return -pi / 6.0
    if abs(x - 2**0.5) < tolerancia_de_erro: return pi / 4.0
    if abs(x + 2**0.5) < tolerancia_de_erro: return -pi / 4.0
    if abs(x - 2*3**0.5/3) < tolerancia_de_erro: return pi / 3.0
    if abs(x + 2*3**0.5/3) < tolerancia_de_erro: return -pi / 3.0

    if abs(x) < 1.0 - tolerancia_de_erro:
        raise ValueError("Arco cossecante indefinido, o domínio dessa função é |x| ≥ 1.")
    return arcsin(1.0 / x)


#Arco tangente de dois argumentos
def arctan2(y: float, x: float) -> float:

    if x == 0 and y == 0:
        return 0.0
    if x == 0:
        return pi / 2.0 if y > 0 else -pi / 2.0
    angulo = arctan(y / x)
    if x < 0:
        angulo += pi if y >= 0 else -pi
    return angulo


#Funções hiperbólicas
#Seno hiperbólico
def sinh(x: float) -> float:

    if abs(x) < tolerancia_de_erro: return 0.0

    from aritmetica import exponencial_e_local
    ex = exponencial_e_local(x)
    enx = exponencial_e_local(-x)
    return (ex - enx) / 2.0


#Cosseno hiperbólico
def cosh(x: float) -> float:

    if abs(x) < tolerancia_de_erro: return 1.0

    from aritmetica import exponencial_e_local
    ex = exponencial_e_local(x)
    enx = exponencial_e_local(-x)
    return (ex + enx) / 2.0

#Tangente hiperbólica
def tanh(x: float) -> float:

    if abs(x) < tolerancia_de_erro: return 0.0

    from aritmetica import exponencial_e_local
    e2x = exponencial_e_local(2.0 * x)
    return (e2x - 1.0) / (e2x + 1.0)


#Cotangente hiperbólica
def coth(x: float) -> float:

    if abs(x) < tolerancia_de_erro:
        raise ZeroDivisionError("Cotangente hiperbólica de 0 é indefinida.")

    from aritmetica import exponencial_e_local
    e2x = exponencial_e_local(2.0 * x)
    return (e2x + 1.0) / (e2x - 1.0)

#Secante hiperbólica
def sech(x: float) -> float:

    if abs(x) < tolerancia_de_erro: return 1.0

    return 1.0 / cosh(x)

#Cossecante hiperbólica
def csch(x: float) -> float:

    if abs(x) < tolerancia_de_erro:
        raise ZeroDivisionError("Cossecante hiperbólica de 0 é indefinida.")

    return 1.0 / sinh(x)
```
