# Números Complexos

  Esse módulo foi desenvolvido para possibilitar a interpretação computacional de números complexos, além de possibilitar operações e transformações dos mesmos. Para isso, foram importadas funções e constantes dos módulos matemáticos desenvolvidos "constantes.py" e "aritmetica.py".

  Deste modo, foi criada uma nova classe chamada "Complexo", essa é uma representação dos números complexos. Nela, foram definidas as partes que compõe o número imaginário na classe, sendo a parte real e a imaginária. Também foram definidas as maneiras com que o número complexo é interpretado internamente, como ele aparecerá ao ser printado, como aparecerá internamente no código, como funciona sua comparação entre números complexos e com números reais. Além disso, foi definido um algoritmo que permite converter inteiros e floats para números complexos. 

  Foram implementadas as funções que realizam operações com números complexos, tanto entre dois números complexos, quanto entre números complexos e reais. Essas operações são de adição, subtração, multiplicação, divisão e negação. Posteriormente, foram feitos códigos que realizam operação operações mais complexas, sendo de potenciação, radiciação, exponenciação na base "e" e o logaritmo natural. 

  Também foram feitos algoritmos para certas propriedades dos números complexos, que envolvem outras operações. Essas foram módulo, argumento, complexo conjugado e o quadrado da norma; também algoritmos que verificam se um número é real e se um número é puramente complexo. Houve também a criação de algoritmos que converter um número complexo para sua forma polar e outro que gera um número complexo a partir de sua forma polar.

  Para finalizar, adicionou-se a linha de código "%%writefile numeros_complexos.py" para converter o código em um módulo exportável para ser usado em outros módulos e no uso final. Pelo mesmo motivo, o código teve que ser colocado inteiramente em uma única célula de código, pois o comando usado para isso exporta códigos de apenas uma célula para isso.

```python
%%writefile numeros_complexos.py

from constantes import pi, tolerancia_de_erro
from aritmetica import valor_absoluto, potencia_rapida_int, piso

#Definição de classe para números complexos
class Complexo:

    __slots__ = ("real", "imag")

    def __init__(proprio, real: float = 0.0, imag: float = 0.0):
        proprio.real = float(real)
        proprio.imag = float(imag)

    def __str__(proprio ) -> str:
        if proprio.imag == 0:
            return f"{proprio.real: .6g}"
        if proprio.real == 0:
            return f"{proprio.imag}i"
        sinal = "+" if proprio.imag >= 0 else "-"
        return f"{proprio.real} {sinal} {valor_absoluto(proprio.imag)}i"

    def __repr__(proprio) -> str:
        return f"Complexo({proprio.real}, {proprio.imag})"

    def __eq__(proprio, outro) -> bool:
        if isinstance(outro, Complexo):
            return (valor_absoluto(proprio.real - outro.real) < tolerancia_de_erro and
                    valor_absoluto(proprio.imag - outro.imag) < tolerancia_de_erro)
        if isinstance(outro, (int, float)):
            return (valor_absoluto(proprio.real - outro) < tolerancia_de_erro and
                    valor_absoluto(proprio.imag) < tolerancia_de_erro)
        return NotImplemented

    
    def __hash__(self):
        return hash((round(self.real / tolerancia_de_erro), round(self.imag / tolerancia_de_erro)))
        

#Conversão de int e float para os complexos
def para_complexo(valor) -> Complexo:
    if isinstance(valor, Complexo):
        return valor
    if isinstance(valor, (int, float)):
        return Complexo(float(valor), 0.0)
    raise TypeError(f"Não é possível converter {type(valor).__name__} para Complexo.")


#Operações
#Adição
def adicao_complexa(a: Complexo, b: Complexo) -> Complexo:
    a, b = para_complexo(a), para_complexo(b)
    return Complexo(a.real + b.real, a.imag + b.imag)


#Subtração
def subtracao_complexa(a: Complexo, b: Complexo) -> Complexo:
    a, b = para_complexo(a), para_complexo(b)
    return Complexo(a.real - b.real, a.imag - b.imag)


#Multiplicação
def multiplicacao_complexa(a: Complexo, b: Complexo) -> Complexo:
    a, b = para_complexo(a), para_complexo(b)
    return Complexo(
        a.real * b.real - a.imag * b.imag,
        a.real * b.imag + a.imag * b.real
    )


#Divisão
def divisao_complexa(a: Complexo, b: Complexo) -> Complexo:
    a, b = para_complexo(a), para_complexo(b)
    denominador = b.real * b.real + b.imag * b.imag
    if denominador < tolerancia_de_erro:
        raise ZeroDivisionError("Divisão por zero não é definida.")
    return Complexo(
        (a.real * b.real + a.imag * b.imag) / denominador,
        (a.imag * b.real - a.real * b.imag) / denominador
    )


#Negação
def negacao_complexa(z: Complexo) -> Complexo:
    z = para_complexo(z)
    return Complexo(-z.real, -z.imag)


#Propriedades
#Módulo
def modulo(z: Complexo) -> float:
    
    from funcoes import sqrt
    
    z = para_complexo(z)
    return sqrt(z.real * z.real + z.imag * z.imag)


#Argumento
def argumento(z: Complexo) -> float:
    
    from trigonometria import arctan2
    
    z = para_complexo(z)
    if z.real == 0 and z.imag == 0:
        return 0.0
    return arctan2(z.imag, z.real)


#Complexo conjugado
def complexo_conjugado(z: Complexo) -> Complexo:
    z = para_complexo(z)
    return Complexo(z.real, -z.imag)


#Norma ao quadrado
def norma_ao_quadrado(z: Complexo) -> float:
    z = para_complexo(z)
    return z.real * z.real + z.imag * z.imag


#Verificação se é um número real
def eh_real(z: Complexo) -> bool:
    z = para_complexo(z)
    return valor_absoluto(z.imag) < tolerancia_de_erro


#Verificação se é puramente imaginário
def eh_puramente_imaginario(z: Complexo) -> bool:
    z = para_complexo(z)
    return (valor_absoluto(z.real) < tolerancia_de_erro and
            valor_absoluto(z.imag) >= tolerancia_de_erro)


#Forma polar
#Conversão para a forma polar
def para_polar(z: Complexo) -> tuple:
    return modulo(z), argumento(z)


#Número complexo a partir da forma polar
def de_polar(r: float, teta: float) -> Complexo:
    
    from trigonometria import sin, cos
    
    if r < 0:
        raise ValueError("O módulo deve ser ≥ 0.")
    return Complexo(r * cos(teta), r * sin(teta))


#Potência complexa
def potencia_complexa(z: Complexo, n: float) -> Complexo:
    
    from trigonometria import sin, cos
    from funcoes import exponencial, ln

    z = para_complexo(z)

    if z.real == 0 and z.imag == 0:
        if n == 0:
            return Complexo(1.0, 0.0)
        if n < 0:
            raise ZeroDivisionError("0 elevado a potência negativa é indefinido.")
        return Complexo(0.0, 0.0)

    if n == 0:
        return Complexo(1.0, 0.0)
    if n == 1:
        return Complexo(z.real, z.imag)

    r, teta = para_polar(z)

    if isinstance(n, int) or (isinstance(n, float) and n == piso(n)):
        n_int = int(n)
        if n_int >= 0:
            r_n = potencia_rapida_int(r, n_int)
        else:
            r_n = 1.0 / potencia_rapida_int(r, -n_int)
    else:
        r_n = exponencial(n * ln(r))

    n_teta = n * teta
    return Complexo(r_n * cos(n_teta), r_n * sin(n_teta))


#Exponenciação na base e
def exponencial_complexa(z: Complexo) -> Complexo:

    from funcoes import exponencial
    from trigonometria import sin, cos
    
    z = para_complexo(z)
    r = exponencial(z.real)
    return Complexo(r * cos(z.imag), r * sin(z.imag))


#Logaritmo natural
def ln_complexo(z: Complexo) -> Complexo:
    
    from funcoes import ln
    
    z = para_complexo(z)
    if z.real == 0 and z.imag == 0:
        raise ValueError("ln(x) é indefinido para x <= 0.")
    return Complexo(ln(modulo(z)), argumento(z))


#Raíz quadrada
def raiz_quadrada_complexa(z: Complexo) -> Complexo:
    
    from funcoes import sqrt
    
    z = para_complexo(z)
    r, teta = para_polar(z)
    return de_polar(sqrt(r), teta / 2.0)
```
