# Constantes matemáticas

  Inicialmente, foi necessário desenvolver uma base para algumas constantes matemáticas. Deste modo, foram criados algoritmos para calcular as constantes pi, número de Euler, phi, e epsilon. Com isso, criou-se com a base do python, além dessas constantes, uma base para introduzir a constante tau (dobro de pi).
  Outros recursos matemáticos definidos foram infinito, menos infinito, e nan (not a number), que servem de base para cálculos mais avançados.
  
  Para definir pi, foi usada a aproximação pela série de Machin; e para isso, usou-se série de arctan via Gregory-Leibniz internamente. 
  Para definir o número de Euler, calculou-se com a aproximação pela série de Taylor. 
  Já para phi, o calculo foi aproximado pela razão áurea; e para isso aproximou-se a raiz quadrada de 5 via método de Newton-Raphson. 
  Por fim, para episilon, foi desenvolvida um algoritmo que determina o épsilon de máquina: menor δ > 0 tal que 1.0 + δ != 1.0. Este valor que representa a precisão relativa do ponto flutuante IEEE 754.

  Em seguida, foram definidas as constantes globais do módulo para as nomear corretamente. e definidas como constantes números que são muito usados e precisam ser calculados toda vez que forem usados, como ln(2) e ln(10).

  Por fim, foram desenvolvidos a número de tolerância de erro e o de máximo de interações. O número de tolerância de erro é para existir uma precisão padrão para comparações numéricas. Já o número máximo de interações é usado para manter valores próximos de itens com muitas ou infinitas casas decimais, limitando a um número que não seja calculado indefinidamente.

  Para finalizar, adicionou-se a linha de código "%%writefile constantes.py" para converter o código em um módulo exportável para ser usado em outros módulos e no uso final. Pelo mesmo motivo, o código teve que ser colocado inteiramente em uma única célula de código, pois o comando usado para isso exporta códigos de apenas uma célula para isso.

```python
%%writefile constantes.py

#Tolerância de erro
tolerancia_de_erro: float = 1e-10

#Número máximo de interações
maximo_de_interacoes: int = 10000


#Pi
def computa_pi(termos: int = 1000) -> float:

    def serie_arctan(x: float, n: int) -> float:
        x2 = x * x
        termo = x
        resultado = 0.0
        sinal = 1
        for k in range(termos):
            resultado += sinal * termo / (2 * k + 1)
            termo *= x2
            sinal = -sinal
            if abs(termo) / (2 * k + 3) < tolerancia_de_erro * 1e-8:
                break
        return resultado

    return 4.0 * (4 * serie_arctan(1 / 5, termos) - serie_arctan(1 / 239, termos))

#Número de Euler
def computa_e(termos: int = 50) -> float:

    resultado = 0.0
    fatorial = 1
    for k in range(termos):
        if k > 0:
            fatorial *= k
        resultado += 1.0 / fatorial
    return resultado

#Phi
def computa_phi() -> float:

    x = 2.0
    for _ in range(100):
        x_new = (x + 5.0 / x) / 2.0
        if abs(x_new - x) < 1e-15:
            break
        x = x_new
    return (1.0 + x) / 2.0

#Epsilon
def computa_epsilon() -> float:

    epsilon = 1.0
    while 1.0 + epsilon / 2.0 != 1.0:
        epsilon /= 2.0
    return epsilon


#Constantes globais

pi: float = computa_pi()

e: float = computa_e()

phi: float = computa_phi()

tau: float = 2.0 * pi

inf: float = float("inf")

neg_inf: float = float("-inf")

nan: float = float("nan")

epsilon: float = computa_epsilon()

#Números usuais

ln2 = 0.6931471805599453

ln10 = 2.302585092994046
```
