# Operações Aritméticas

  Para desenvolver a base aritmética, desenvolveu-se algoritmos para calcular operações e gerar códigos que servem de base para funções em outros módulos. As operações aqui foram transformadas em algoritmos usando as próprias operações do Python e a lógica matemática por trás dessas operações
  
  As operações desenvolvidas foram adição, subtração, multiplicação, divisão, potenciação, radiciação, valor absoluto e fatorial. Além disso, há alguns algoritmos de análise numérica, tal como o resto da divisão, mínimo multiplo comum, máximo divisor comum, fatoração, contagem de digitos, soma de digitos, análise do sinal e para verificar se um número é primo. E alguns algoritmos que servem de auxiliares para outros algoritmos, como o piso, teto e potenciação e logaritmo locais com bases e.

  Para finalizar, adicionou-se a linha de código "%%writefile aritmetica.py" para converter o código em um módulo exportável para ser usado em outros módulos e no uso final. Pelo mesmo motivo, o código teve que ser colocado inteiramente em uma única célula de código, pois o comando usado para isso exporta códigos de apenas uma célula para isso.

```python
%%writefile aritmetica.py

from constantes import tolerancia_de_erro, maximo_de_interacoes

#Adição
def adicao(a: float, b: float) -> float:
    return a + b


#Subtração
def subtracao(a: float, b: float) -> float:
    return a - b


#Multiplicação
def multiplicacao(a: float, b: float) -> float:
    return a * b


#Divisão
def divisao(a: float, b: float) -> float:

    if b == 0:
        raise ZeroDivisionError("Divisão por zero não é definida.")
    return a / b


#Resto da divisão
def resto_divisao(a: float, b: float) -> float:

    if b == 0:
        raise ZeroDivisionError("Módulo por zero não é definido.")
    return a - b * piso(a / b)


#Piso
def piso(x: float) -> int:

    n = int(x)

    if x < 0 and x != n:
        return n - 1
    return n


#Teto
def teto(x: float) -> int:

    return -piso(-x)


#Potência
def potencia(base: float, exp: float) -> float:

    if exp == 0:
        return 1.0

    if base == 0:
        if exp > 0:
            return 0.0
        raise ZeroDivisionError("0 elevado a potência negativa é indefinido.")

    if isinstance(exp, int) or (isinstance(exp, float) and exp == int(exp)):
        exp_int = int(exp)
        if exp_int < 0:
            return 1.0 / potencia_rapida_int(base, -exp_int)
        return potencia_rapida_int(base, exp_int)

    if base < 0:
        raise ValueError("Potência fracionária de base negativa resulta em número complexo (usar operações complexas).")

    return exponencial_e_local(exp * ln_local(base))


#Potência rápida
def potencia_rapida_int(base: float, exp: int) -> float:

    if exp == 0:
        return 1.0
    if exp == 1:
        return base
    if exp % 2 == 0:
        metade = potencia_rapida_int(base, exp // 2)
        return metade * metade
    return base * potencia_rapida_int(base, exp - 1)


#logaritmo natural local
def ln_local(x: float) -> float:

    if x <= 0:
        raise ValueError("ln(x) é indefinido para x <= 0.")
    if x == 1:
        return 0.0

    k = 0
    m = x
    while m > 2.0:
        m /= 2.0
        k += 1
    while m < 0.5:
        m *= 2.0
        k -= 1

    y = (m - 1.0) / (m + 1.0)
    y2 = y * y
    termo = y
    resultado = 0.0
    for n in range(maximo_de_interacoes):
        resultado += termo / (2 * n + 1)
        termo *= y2
        if abs(termo) < tolerancia_de_erro * 1e-5:
            break

    ln2 = 0.6931471805599453
    return 2.0 * resultado + k * ln2


#Potência de base e
def exponencial_e_local(x: float) -> float:

    ln2 = 0.6931471805599453
    if x > 709:
        return float("inf")
    if x < -745:
        return 0.0

    n = int(x / ln2 + 0.5)
    r = x - n * ln2

    termo = 1.0
    resultado = 1.0
    for k in range(1, maximo_de_interacoes):
        termo *= r / k
        resultado += termo
        if abs(termo) < tolerancia_de_erro * 1e-8:
            break

    return resultado * potencia_rapida_int(2.0, n) if n >= 0 else resultado / potencia_rapida_int(2.0, -n)


#Raíz n
def n_raiz(x: float, n: int) -> float:

    if n == 0:
        raise ValueError("Índice de raiz não pode ser zero.")
    if n < 0:
        return 1.0 / n_raiz(x, -n)
    if x == 0:
        return 0.0
    if n == 1:
        return float(x)
    if x < 0:
        if n % 2 == 0:
            raise ValueError("Raiz n-ésima de índice par de um número negativo resulta em um número complexo (usar operações complexas).")
        return -n_raiz(-x, n)

    estimativa = x / n if x > 1 else x + 0.1

    for _ in range(maximo_de_interacoes):
        r_pow = potencia_rapida_int(estimativa, n - 1)
        if r_pow == 0:
            break
        nova_estimativa = ((n - 1) * estimativa + x / r_pow) / n
        if abs(nova_estimativa - estimativa) < tolerancia_de_erro:
            return nova_estimativa
        estimativa = nova_estimativa

    return estimativa


#Valor absoluto
def valor_absoluto(x: float) -> float:

    return x if x >= 0 else -x


#Sinal
def sinal(x: float) -> int:

    if x > 0:
        return 1
    if x < 0:
        return -1
    return 0


#Teoria dos números

#Máximo divisor comum
def mdc(a: int, b: int) -> int:

    a, b = valor_absoluto(int(a)), valor_absoluto(int(b))
    while b:
        a, b = b, a % b
    return a


#Mínimo multiplo comum
def mmc(a: int, b: int) -> int:

    if a == 0 or b == 0:
        return 0
    return valor_absoluto(int(a) * int(b)) // mdc(a, b)


#Fatorial
def fatorial(n: int) -> int:

    if not isinstance(n, (int, float)) or (isinstance(n, float) and n != int(n)):
        raise TypeError("Fatorial só opera em números inteiros.")
    n = int(n)
    if n < 0:
        raise ValueError("Fatorial de número negativo é indefinido.")
    resultado = 1
    for k in range(2, n + 1):
        resultado *= k
    return resultado


#Verificação se um número é primo ou não
def eh_primo(n: int) -> bool:

    n = int(n)
    if n < 2:
        return False
    if n == 2 or n == 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False

    k = 5
    while k * k <= n:
        if n % k == 0 or n % (k + 2) == 0:
            return False
        k += 6
    return True


#Fatoração em números primos
def fatores_primos(n: int) -> list:

    n = int(valor_absoluto(n))
    if n < 2:
        return []
    fatores = []

    while n % 2 == 0:
        fatores.append(2)
        n //= 2

    d = 3
    while d * d <= n:
        while n % d == 0:
            fatores.append(d)
            n //= d
        d += 2
    if n > 1:
        fatores.append(n)
    return fatores


#Contagem de digitos
def contagem_digitos(n: int) -> int:

    n = valor_absoluto(int(n))
    if n == 0:
        return 1
    contagem = 0
    while n > 0:
        contagem += 1
        n //= 10
    return contagem


#Soma dos digitos
def soma_digitos(n: int) -> int:

    n = valor_absoluto(int(n))
    total = 0
    while n > 0:
        total += n % 10
        n //= 10
    return total
```
